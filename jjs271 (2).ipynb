```python
import pandas as pd
import numpy as np
%matplotlib inline
import geopandas
import matplotlib.pyplot as plt

neighborhoods = geopandas.read_file("Neighborhoods/Neighborhoods_.shp")
neighborhoods.plot()
```

```python
fireDistricts = geopandas.read_file("FireDistricts/Fire_Districts.shp")
fireDistricts = fireDistricts.to_crs(neighborhoods.crs)
fireDistricts.plot()
```

```python
facilities = pd.read_csv("Facilities.csv", 
                         usecols=["name","longitude","latitude","neighborhood"])
facilities.iloc[100:127]
```

```python
x = facilities.iloc[100:127]["longitude"]
y = facilities.iloc[100:127]["latitude"]
plt.scatter(x, y)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
neighborhoods.plot(ax=ax, color="black", edgecolor="black", cmap="Oranges")
fireDistricts.plot(ax=ax, color="orange", edgecolor="black", alpha=.7)
plt.scatter(x=facilities.iloc[100:127]["longitude"], y=facilities.iloc[100:127]["latitude"], color="red", marker="o")
plt.show()
```
